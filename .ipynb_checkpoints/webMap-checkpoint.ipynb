```python
import pandas as pd
import geopandas as gpd
import os
from shapely.geometry import LineString
import matplotlib.pyplot as plt

os.chdir(r'c:\users\brenn\documents\projects\HopSkipJump')
```

```python
gdf = gpd.read_file('Breweries/DenverBreweries')
gdf.reset_index(inplace = True, names = 'id')
```

```python
gdf_point = gdf.copy()

gdf_point['lat'] = gdf_point.geometry.y
gdf_point['lon'] = gdf_point.geometry.x

gdf_point = gdf_point[['id', 'lat', 'lon']].copy()

gdf_point.to_csv('Breweries/brew_poi.csv', index = False)
```

```python
ttm = pd.read_csv('Breweries/brew_matrix.csv')
ttm.drop(columns = "Unnamed: 0", inplace = True)
```

```python
jgdf = gdf[['id', 'Name', 'geometry']].copy()

ndf = ttm.merge(jgdf, left_on = 'from_id', right_on = 'id', how = 'inner').merge(jgdf, left_on = 'to_id', right_on = 'id', how = 'inner', suffixes = ('_from', '_to'))

ndf['geometry'] = ndf.apply(lambda r: LineString([r.geometry_from, r.geometry_to]), axis = 1)

ndf = gpd.GeoDataFrame(ndf, crs = gdf.crs)
```

```python
ndf.query('travel_time_p50 <= 15 and from_id != to_id', inplace = True)
```

```python
ndf.sort_values("travel_time_p50", ascending = True, inplace = True)
ndf = ndf.groupby('from_id').head(5).sort_values(['from_id', 'travel_time_p50'], ascending = True)
```

ndf is line plot

gdf is point plot

```python
ndf_merge = ndf[['from_id', 'to_id']].copy()
ndf_merge.head()
```

```python
lister = lambda x: ','.join([str(i) for i in list(set(x))])
ndf_merge = ndf_merge.groupby('from_id').agg({'to_id' : lister}).reset_index()
```

```python
ndf_merge
```

```python
clist = ['#bababa', # General
        '#404040', # Line Link
        '#f4a582', # Connected
        '#ca0020'] # Active
```

```python
m = gdf.explore(color = clist[0], tiles = 'Stamen Toner',  marker_kwds = {'radius' : 7}, style_kwds = {'edgecolor' : 'k'})
ndf.query('from_id == 2').explore(m=m, color = clist[1], style_kwds = {'linewidth' : 10})
ql = ndf_merge.query('from_id == 2').to_id.values[0].split(',')
ql = [int(i) for i in ql]

gdf.query('id in @ql').explore(m=m, color = clist[2], marker_kwds = {'radius' : 7}, style_kwds = {'edgecolor' : 'k'})
gdf.query('id == 2').explore(m=m, color = clist[3], marker_kwds = {'radius' : 7}, style_kwds = {'edgecolor' : 'k'})
m
```
